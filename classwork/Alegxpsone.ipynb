```python
import numpy as np
import random
import math
```

```python
loc = np.loadtxt("/mnt/data/locators.csv", delimiter=",")
dist = np.loadtxt("/mnt/data/approx_dists.csv", delimiter=",")
```

```python
x = loc[:,0]
y = loc[:,1]
d = dist
```

```python
x1, y1, d1 = x[0], y[0], d[0]
```

```python
A = []
b = []

for i in range(1, len(x)):
    A.append([
        2*(x[i] - x1),
        2*(y[i] - y1)
    ])
    
    b.append(
        d1**2 - d[i]**2 + x[i]**2 - x1**2 + y[i]**2 - y1**2
    )

A = np.array(A)
b = np.array(b)
```

```python
p, _, _, _ = np.linalg.lstsq(A, b, rcond=None)
```

```python
bx, by = p
```

```python
print("Battleship coordinates:", bx, by)
```
