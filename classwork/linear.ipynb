```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

a = 0
b = 8
n = 50
#x_points = np.linspace(a, b, n)
x_points = np.random.uniform(a, b, n)
x_points
```

```python
def test_func(x):
    return 2 * x + 1
```

```python
noise_level = 0.5
random_noise = np.random.uniform(-noise_level, noise_level, n)
y_points = test_func(x_points) + random_noise
y_points
```

```python
fig,ax = plt.subplots()
test_y = test_func(x_points)
ax.scatter(x_points, y_points)
ax.plot(x_points, test_y, label='True Function', color='red')
```

```python
a11 = np.sum(x_points)
a12 = np.sum([x**2 for x in x_points   ])
a21 = n
a22 = np.sum(x_points)
A = np.array([[a11, a12], 
            [a21, a22]])
A
```

```python
y1 = np.sum([y_points[i] * x_points[i] for i in range(n)])
y2 = np.sum(y_points)
y = np.array([y1, y2])
```

```python
c = np.linalg.solve(A, y)
c
```

```python
res_y = [c[1] + c[0]*x for x in x_points]
```

```python
fig, ax = plt.subplots()
test_y = test_func(x_points)
ax.scatter(x_points, y_points)
ax.plot(x_points, test_y, label='True Function', color='red')
ax.plot(x_points, res_y, label='Fitted Line', color='green')    
```
