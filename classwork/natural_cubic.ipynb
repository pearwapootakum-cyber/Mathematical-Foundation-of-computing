```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.zeros((8, 8))
b = np.zeros(8)
```

```python
A[0][0] = 1
```

```python
A[1][0] = A[1][1] = A[1][2] = A[1][3] = 1
```

```python
A[2][4] = A[2][5] = A[2][6] = A[2][7] = 1
```

```python
A[3][4] = 1
A[3][5] = 2
A[3][6] = 4
A[3][7] = 8
```

```python
A[4][1] = 1
A[4][2] = 2
A[4][3] = 3
A[4][5] = -1
A[4][6] = -2
A[4][7] = -3
```

```python
A[5][2] = 2
A[5][3] = 6
A[5][6] = -2
A[5][7] = -6
```

```python
A[6][2] = 2
```

```python
A[7][6] = 2
A[7][7] = 12
```

```python
A
```

```python
b =[1, 2, 2, 0, 0, 0, 0, 0]
b
```

```python
res = np.linalg.solve(A, b)
res
```

```python
s0_coeff = res[:4]
s1_coeff = res[4:]
s0_coeff, s1_coeff
```

```python
def poly(c_vect, x):
    res = 0
    for i in range(len(c_vect) ):
        res += c_vect[i] * (x ** i)
    return res
```

```python
x_for_plot_s0 = np.linspace(0, 1, 100)
x_for_plot_s1 = np.linspace(1, 2, 100)

y_for_plot_s0 = []
y_for_plot_s1 = []

for x in x_for_plot_s0:
    y_for_plot_s0.append(poly(s0_coeff, x))

for x in x_for_plot_s1:
    y_for_plot_s1.append(poly(s1_coeff, x))
```

```python
plt.plot(x_for_plot_s0, y_for_plot_s0, color='blue', label='S0')
plt.plot(x_for_plot_s1, y_for_plot_s1, color='orange', label='S1')
plt.scatter([0,1,2], [1,2,0], color='red')

plt.legend()
plt.grid(True)
plt.show()
```
